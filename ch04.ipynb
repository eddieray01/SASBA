## Chapter 4: Characteristics of the Sampling Distribution Mean

```sas
/* Program 4.2*/
data restaurant;

call streaminit(1); /*set random seed;*/
do j = 1 to 100000;                 
   do i = 1 to 100;                 
      time = rand("exponential",20);
      output;
      end;
   end;
run;

proc sort data=restaurant; by j;
run;

proc means noprint; 
   var time;    /*calculate sample mean time for 100,000 samples;*/
   output out=samplingdistr mean=meantime;
   by j;
run;

proc univariate data=samplingdistr;
   histogram meantime;
   inset nobs mean median q1 q3 skewness kurtosis / pos=ne;
   var meantime;
run;
```

# Results
    sampling distribution of 100,000 sample means by taking 100,100 random samples of size 100 from an exponential having mean of 20 minutes
    

```sas
/* Program 4.3 t-Test of BMI */
data patient;
    set SASBA.diab25f;
run;

proc print data=patient(obs=5);
run;
```

```sas
proc ttest data=patient
alpha=.10 h0=30;
var bmi;
title 't-Test';
run;
```

# Result
    Ho: u>=30  Ha: u<30
    test at 0.10 level of significance and hypthoesized mean BMI to be 30 (Uo)
    two-tailed so p-value .0216/2=0.0108
    reject null hypothesis, concluding there is evidence that female diabetics have a BMI <30
